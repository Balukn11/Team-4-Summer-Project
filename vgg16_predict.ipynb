# Loading model

```python
import tensorflow as tf
from keras.models import load_model
model1 = load_model('C:\\Users\\ADMIN\\vggmodelnew.h5')
```

```python
model1.summary()
```

# Prediction

```python
from tensorflow.keras.preprocessing import image
from matplotlib import pyplot as plt
img =image.load_img('G:\\Main\\Train\\Positive\\04658.jpg',target_size=(227,227))
plt.imshow(img)
```

```python
import numpy as np
from tensorflow.keras.applications.vgg16 import preprocess_input
img = image.img_to_array(img)
img = img/255.0
img = np.expand_dims(img,axis=0)
pred = np.argmax(model1.predict(img),axis=1)
print(pred)
```

```python
if pred[0] == 0:
    print("This image contains Crack!!!")
else:
    print("This image doesn't contain Crack!!!")
```
