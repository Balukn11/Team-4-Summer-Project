# Splitting Data as train and test


```python
test_positive = 'G:\\Main\\Test\\Positive'
test_negative = 'G:\\Main\\Test\\Negative'
train_positive = 'G:\\Main\\Train\\Positive'
train_negative = 'G:\\Main\\Train\\Negative'

data_positive = 'G:\\Main\\Dataset\\Positive'
data_negative = 'G:\\Main\\Dataset\\Negative'

print("No.of positive images: " ,len(no_of_postive))
print("No.of negative images: " ,len(no_of_negative))
```

```python
import random
```

```python
import shutil
for f in no_of_postive:
    if random.random() > 0.80:
        shutil.move(f'{data_positive}\\{f}',test_positive )
    else:
        shutil.move(f'{data_positive}\\{f}',train_positive)
```
